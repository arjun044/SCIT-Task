```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
a = np.random.poisson(100.0,(1000))
b = np.random.poisson(100.0,(1000))
c = np.random.poisson(100.0,(1000))
print(a.shape[0])
print(b.shape[0])
print(c.shape[0])
```

```python
plt.figure(figsize=(13,8))
plt.hist(a,alpha=0.5)
plt.hist(b,alpha=0.5)
plt.hist(c,alpha=0.5)
plt.legend(['a','b','c'],bbox_to_anchor=(0,0,0.5,1))
plt.grid()
plt.show()
```

The above graph 

```python
d = np.random.normal(100.0,size=(1000))
e = np.random.normal(100.0,size=(1000))
f = np.random.normal(100.0,size=(1000))
print(d.shape[0])
print(e.shape[0])
print(f.shape[0])
```

```python
plt.figure(figsize=(13,8))
plt.hist(d,alpha=0.5)
plt.hist(e,alpha=0.5)
plt.hist(f,alpha=0.5)
plt.legend(['d','e','f'],bbox_to_anchor=(0,0,0.5,1))
plt.grid()
plt.show()
```
