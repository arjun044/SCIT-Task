```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# generating 1000 samples for 3 times
a = np.random.poisson(100.0,(1000))
b = np.random.poisson(100.0,(1000))
c = np.random.poisson(100.0,(1000))
print(a.shape[0])
print(b.shape[0])
print(c.shape[0])
np.savetxt('SCIT/datapoints_csv/poisson_data_points.csv',np.c_[a,b,c],delimiter=',')
```

```python
plt.hist(a,alpha=0.5,density=True)
plt.legend(['a'],bbox_to_anchor=(0,0,0.5,1))
plt.grid()
plt.xlabel('x')
plt.ylabel('Probability')
plt.savefig("SCIT/plots/a_poisson.png")
plt.show()
```

```python
plt.hist(b,alpha=0.5,density=True)
plt.legend(['b'],bbox_to_anchor=(0,0,0.5,1))
plt.grid()
plt.xlabel('x')
plt.ylabel('Probability')
plt.savefig("SCIT/plots/b_poisson.png")
plt.show()
```

```python
plt.hist(c,alpha=0.5,density=True)
plt.legend(['c'],bbox_to_anchor=(0,0,0.5,1))
plt.grid()
plt.xlabel('x')
plt.ylabel('Probability')
plt.savefig("SCIT/plots/c_poisson.png")
plt.show()
```

```python
plt.hist(a,alpha=0.5,density=True)
plt.hist(b,alpha=0.5,density=True)
plt.hist(c,alpha=0.5,density=True)
plt.legend(['a','b','c'],bbox_to_anchor=(0,0,0.5,1))
plt.grid()
plt.xlabel('x')
plt.ylabel('Probability')
plt.savefig("SCIT/plots/abc_poisson.png")
plt.show()
```

```python
# generating 1000 samples for 3 times
d = np.random.normal(100.0,10,size=(1000))
e = np.random.normal(100.0,10,size=(1000))
f = np.random.normal(100.0,10,size=(1000))
print(d.shape[0])
print(e.shape[0])
print(f.shape[0])
np.savetxt('SCIT/datapoints_csv/normal_data_points.csv',np.c_[d,e,f],delimiter=',')
```

```python
plt.hist(d,alpha=0.5,density=True)
plt.legend(['d'],bbox_to_anchor=(0,0,0.5,1))
plt.grid()
plt.xlabel('x')
plt.ylabel('Probability')
plt.savefig("SCIT/plots/d_normal.png")
plt.show()
```

```python
plt.hist(e,alpha=0.5,density=True)
plt.legend(['e'],bbox_to_anchor=(0,0,0.5,1))
plt.grid()
plt.xlabel('x')
plt.ylabel('Probability')
plt.savefig("SCIT/plots/e_normal.png")
plt.show()
```

```python
plt.hist(f,alpha=0.5,density=True)
plt.legend(['f'],bbox_to_anchor=(0,0,0.5,1))
plt.grid()
plt.xlabel('x')
plt.ylabel('Probability')
plt.savefig("SCIT/plots/f_normal.png")
plt.show()
```

```python
plt.hist(d,alpha=0.5,density=True)
plt.hist(e,alpha=0.5,density=True)
plt.hist(f,alpha=0.5,density=True)
plt.legend(['d','e','f'],bbox_to_anchor=(0,0,0.5,1))
plt.grid()
plt.xlabel('x')
plt.ylabel('Probability')
plt.savefig("SCIT/plots/def_normal.png")
plt.show()
```

```python
# Generating the average of 100 samples from the poisson distribution for 100 experiment without replacement.
a1 = []
a2 = []
a3 = []
for i in range(0,100):
    a1.append(np.sum(np.random.choice(a,size=(100),replace=False))/100)
    a2.append(np.sum(np.random.choice(b,size=(100),replace=False))/100)
    a3.append(np.sum(np.random.choice(c,size=(100),replace=False))/100)
np.savetxt('SCIT/datapoints_csv/poisson_data_points_average.csv',np.c_[a1,a2,a3],delimiter=',')
```

```python
# plotting the average of these 100 experiments for 3 cases -> (a,b,c) of poission disribution 
plt.figure(figsize=(13,8))
plt.hist(a1,alpha=0.5,density=True)
plt.hist(a2,alpha=0.5,density=True)
plt.hist(a3,alpha=0.5,density=True)
plt.legend(['a','b','c'],bbox_to_anchor=(0,0,0.1,1))
plt.grid()
plt.xlabel('x')
plt.ylabel('Probability')
plt.savefig("SCIT/plots/def_poisson_average.png")
plt.show()
```

```python
# Generating the average of 100 samples from the normal distribution for 100 experiment without replacement.
b1 = []
b2 = []
b3 = []
for i in range(0,100):
    b1.append(np.sum(np.random.choice(d,size=(100),replace=False))/100)
    b2.append(np.sum(np.random.choice(e,size=(100),replace=False))/100)
    b3.append(np.sum(np.random.choice(f,size=(100),replace=False))/100)
np.savetxt('SCIT/datapoints_csv/normal_data_points_average.csv',np.c_[b1,b2,b3],delimiter=',')
```

```python
# plotting the average of these 100 experiments for 3 cases -> (a,b,c) of normal disribution 
plt.figure(figsize=(13,8))
plt.hist(b1,alpha=0.5,density=True)
plt.hist(b2,alpha=0.5,density=True)
plt.hist(b3,alpha=0.5,density=True)
plt.legend(['d','e','f'],bbox_to_anchor=(0,0,0.1,1))
plt.grid()
plt.xlabel('x')
plt.ylabel('Probability')
plt.savefig("SCIT/plots/def_normal_average.png")
plt.show()
```
